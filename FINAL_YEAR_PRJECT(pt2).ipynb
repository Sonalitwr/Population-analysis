```python
#dividing data into dependent and independent variable 
x = df_lit.drop("Literacy Rate",axis = 1)
y = df_lit["Literacy Rate"]
```

```python
# removing  skewness 
#importing necessary library
from sklearn.preprocessing import power_transform


#intialize
df1 = power_transform(x_s)

#create the new dataframe

df1 = pd.DataFrame(df1, columns = x_s.columns)

#check skewness
df1.skew()
```

```python
#skewness removed
```

DATA STANDARDIZATION

```python
#scaling standardized the model so that accuracy is higher 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x_s = scale.fit_transform(x_s) 
```

```python
# finding best random state

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
max_accu = 0
max_rs = 0
for i in range(0,200):
    x_train,x_test,y_train,y_test = train_test_split(x_s,y,test_size = .10, random_state = i)
    LR = LinearRegression()
    LR.fit(x_train, y_train)
    predf = LR.predict(x_test)
    acc = r2_score(y_test, predf)
    if acc>max_accu:
        max_accu = acc
        max_rs = i
print("Accuracy= ", max_accu, "  Random State= ", max_rs)
```

Checking algorithms 

```python
# LINEAR REGRESSION

from sklearn.metrics import mean_squared_error, mean_absolute_error
x_train,x_test,y_train,y_test = train_test_split(x_s,y,test_size = 10 , random_state = 177)
LR = LinearRegression()
LR.fit(x_train, y_train)
pred = LR.predict(x_test)

print(r2_score(y_test, pred))
print("Mean absolute error: ", mean_absolute_error(y_test,pred))
print("Mean squared error: ", mean_squared_error(y_test,pred))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, pred)))
```

```python
# KNeighborsRegressor

from sklearn.neighbors import KNeighborsRegressor
knnr = KNeighborsRegressor(n_neighbors = 20)
knnr.fit(x_train, y_train)
pred2 = knnr.predict(x_test)

print(r2_score(y_test,pred2))


print("Mean absolute error: ", mean_absolute_error(y_test,pred2))
print("Mean squared error: ", mean_squared_error(y_test,pred2))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, pred2)))
```

```python
# DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor()
DTR.fit(x_train,y_train)
pred3 = DTR.predict(x_test)
print(r2_score(y_test,pred3))

print("Mean absolute error: ", mean_absolute_error(y_test,pred3))
print("Mean squared error: ", mean_squared_error(y_test,pred3))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, pred3)))
```

```python
# AdaBoostRegressor

from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import cross_val_score, KFold

ada_reg = AdaBoostRegressor(n_estimators = 100)
ada_reg.fit(x_train,y_train)
pred4 = ada_reg.predict(x_test)

print(r2_score(y_test,pred4))


print("Mean absolute error: ", mean_absolute_error(y_test,pred4))
print("Mean squared error: ", mean_squared_error(y_test,pred4))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, pred4)))
```
